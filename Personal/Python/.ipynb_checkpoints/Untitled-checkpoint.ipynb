```python
'''
Fixed Income Asset Allocator:
For each ratings based fund, calculate:
    OAS relative to JUCO
    MA of OAS relative to JUCO over t months
    Standard deviation of MA of OAS relative to JUCO over t months
    Z-score of each current relative OAS to MA of relative OAS: (thisMonthRelativeSpread - MA RelativeSpread) / SDev of relative spread over MA
    momentum signal
'''
import math
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
bondData = pd.read_csv('/Users/jonahschwartzman/Downloads/computerScience/Personal/Python/rawData.csv')
```

```python
print(bondData.columns)
print(bondData.shape)
print(bondData.describe)
```

```python
dates, xbbMTR, xbMTR, xcccMTR, aggMTR, xbbOAS, xbOAS, xcccOAS, aggOAS, xbbMV, xbMV, xcccMV, aggMV, xbbW, xbW, xcccW = [bondData[col] for col in bondData.columns ][1:]
```

```python
xcccRelativeSpreads, xbRelativeSpreads, xbbRelativeSpreads = [oas / aggOAS for oas in [xcccOAS, xbOAS, xbbOAS]]

totalMonths = len(aggOAS)
t = 96

xcccAvgRelativeSpreads, xbAvgRelativeSpreads, xbbAvgRelativeSpreads = [[np.average(relativeSpreads[i:i+t]) for i in range(0, totalMonths-t)] for relativeSpreads in [xcccRelativeSpreads, xbRelativeSpreads, xbbRelativeSpreads]]

xcccStds, xbStds, xbbStds = [[np.std(relativeSpreads[i:i+t]) for i in range(0, totalMonths-t)] for relativeSpreads in [xcccRelativeSpreads, xbRelativeSpreads, xbbRelativeSpreads]]

xcccZScores, xbZScores, xbbZScores = [[(relativeSpreads[i+t] - relativeSpreads[i]) / stds[i] for i in range(0, totalMonths-t)] for relativeSpreads, stds in [(xcccRelativeSpreads, xcccStds), (xbRelativeSpreads, xbStds), (xbbRelativeSpreads, xbbStds)]]
```

```python
plt.plot(xcccOAS, label = 'XCCC', color = 'b')
plt.plot(xbbOAS, label = 'XBB', color = 'g')
plt.plot(xbOAS, label = 'XB', color = 'r')
plt.plot(aggOAS, label = 'Aggregate Spread', c = 'k')
plt.xlabel("Months since 12/31/96")
plt.ylabel("Fund OAS")
plt.title("Fund Spreads")
plt.legend()

plt.show()
```

```python
xAxis = np.array(range(totalMonths))
plt.plot(xAxis, xcccRelativeSpreads, label = "XCCC Relative Spreads", c = 'b')
plt.plot(xAxis, xbbRelativeSpreads, label = "XBB Relative Spreads", c = 'g')
plt.plot(xAxis, xbRelativeSpreads, label = "XB Relative Spreads", c = 'r')
plt.plot(np.array(range(totalMonths)), np.ones(totalMonths), marker = ',', c = (0, 0, 0, .2), label = 'Base')


avgXAxis = np.array(range(t, totalMonths))      
plt.plot(avgXAxis, xcccAvgRelativeSpreads, label = "XCCC Average Relative Spreads", c = (0, 0, 1, .4))
plt.plot(avgXAxis, xbbAvgRelativeSpreads, label = "XBB Average Relative Spreads", c = (0, 1, 0, .4))
plt.plot(avgXAxis, xbAvgRelativeSpreads, label = "XB Average Relative Spreads", c = (1, 0, 0, .4))

plt.legend()
plt.xlabel("Months since 12/31/1996")
plt.ylabel("Fund Relative Spreads")
plt.title("Relative Spreads to JUCO")
plt.show()
```

```python
plt.plot(avgXAxis, xcccStds, label = 'XCCC', c = 'b')
plt.plot(avgXAxis, xbbStds, label = 'XBBB', c = 'g')
plt.plot(avgXAxis, xbStds, label = 'XB', c = 'r')
plt.legend()
plt.xlabel("Months since 12/31/96")
plt.ylabel("Standard Deviation")
plt.title(f"Standard Deviation of Spread Over Past {t} Months")
plt.show()
```

```python
plt.plot(xcccZScores, label = 'XCCC')
plt.plot(xbbZScores, label = 'XBB')
plt.plot(xbZScores, label = 'XB')
#plt.plot([xcccZScores[i] + xbbZScores[i] + xbZScores[i] for i in range(len(xcccZScores))])
plt.legend()
plt.xlabel("Months since 12/31/1996")
plt.ylabel("Z-Score")
plt.title(f"Z-scores of each month over past {t} months")
plt.show()
```
